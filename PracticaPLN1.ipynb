Primero tomaremos un sitio wed y analizaremos el texto para cer de que trata la paguina wed

la libreria urlib ayuda a obeer los datos de la paguina wed.

```python
import urllib.request
response = urllib.request.urlopen('https://es.wikipedia.org/wiki/SpaceX')
html = response.read()
print(html)
```

Utilizamos Beatiful soup que es una libreria en python para eztraer datos de archivos HTML y XML. Utilizamos esta libreria para limpiar el texto de la pagina wed de las etiquetas.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html5lib')
text = soup.get_text(strip=True)
print(text)
```

Una vez tenemos el texto plano que obtuvimos de la pagiina web, vamos a combertir este texto a tokens:

```python
tokens = [t for t in text.split()]
print(tokens)
```

# Cuenta la frecuencia de las palabras

NTKLofrece una funcion FreqDist() que hace el trabajo por nosotros. Ademas de esto se eliminan las palabras(a, at, the, for, etc), ya que 
Finallmente, trazamos el grafico de las palabras que aparece con mas frecuencia en la paguina web.

Finalmente, trazamos el grafico de las palabras que aparecen mas frecuencia en la paguina web.

```python
from nltk.corpus import stopwords
sr = stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
    print(str(key)+':'+str(val))
freq.plot(20, cumulative = False)
```
